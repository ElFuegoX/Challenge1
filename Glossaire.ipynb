# Glossaire des variables pour le challenge ANIP

Ce glossaire documente les variables des quatre fichiers CSV (`demographic_data.csv`, `social_data.csv`, `fdi_data.csv`, `gdp_data.csv`) pour analyser l'attractivité économique, démographique et sociale du Bénin et des régions associées. Il inclut les noms, définitions, unités, sources confirmées, périodes et niveaux géographiques, facilitant la transparence, l’analyse  et les visualisations Power BI.

## 1. Fichier : demographic_data.csv
**Description générale** : Données démographiques pour le Bénin et l'Afrique de l'Ouest, incluant population, natalité, mortalité et migrations.

| Nom de la variable | Définition | Unité | Source confirmée | Période | Géographie |
|--------------------|------------|-------|------------------|---------|------------|
| Country | Nom du pays ou de la région | Aucune | ONU / World Bank | 1950-2024 | Bénin, Afrique de l'Ouest |
| Year | Année de référence pour les données démographiques | Aucune | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| PopTotal | Population totale | Milliers d'habitants | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| PopDensity | Densité de population | Habitants par km² | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| PopSexRatio | Ratio de masculinité (hommes pour 100 femmes) | Hommes/100 femmes | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| MedianAgePop | Âge médian de la population | Années | ONU / World Bank | 1950-2024 | Bénin, Afrique de l'Ouest |
| CBR | Taux brut de natalité (Crude Birth Rate) | Naissances pour 1000 habitants | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| TFR | Taux de fécondité total (Total Fertility Rate) | Enfants par femme | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| LEx | Espérance de vie à la naissance | Années | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| IMR | Taux de mortalité infantile (Infant Mortality Rate) | Décès pour 1000 naissances vivantes | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| Q5 | Taux de mortalité des moins de 5 ans | Décès pour 1000 naissances vivantes | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |
| NetMigrations | Solde migratoire net | Milliers de personnes | UN WPP | 1950-2024 | Bénin, Afrique de l'Ouest |

## 2. Fichier : social_data.csv
**Description générale** : Indicateurs sociaux pour le Bénin et l'Afrique subsaharienne, en format large (séries temporelles pour HDI, GDI, etc.), issus du Human Development Report.

| Nom de la variable | Définition | Unité | Source confirmée | Période | Géographie |
|--------------------|------------|-------|------------------|---------|------------|
| country | Nom du pays ou de la région (renommé "Country") | Aucune | UNDP (Human Development Report) | 1990-2023 | Bénin, Afrique subsaharienne |
| hdi_[year] | Human Development Index (value) | Aucune  | UNDP (Human Development Report) | 1990-2023 | Bénin, Afrique subsaharienne |
| mys_[year] | Mean Years of Schooling (years) | Années | UNDP (Human Development Report) | 1990-2023 | Bénin, Afrique subsaharienne |
| gdi_[year] | Gender Development Index (value) |  Aucune  | UNDP (Human Development Report) | 1990-2023 | Bénin, Afrique subsaharienne |
| gnipc_[year] | Gross National Income Per Capita (2021 PPP$) | USD (PPP, parité de pouvoir d'achat) | UNDP (Human Development Report) | 1990-2023 | Bénin, Afrique subsaharienne |
| gii_[year] | Gender Inequality Index (value) | Aucune | UNDP (Human Development Report) | 1990-2023 | Bénin, Afrique subsaharienne |

## 3. Fichier : fdi_data.csv
**Description générale** : Données sur les Investissements Directs Étrangers (FDI) en flux et stocks, entrants et sortants, pour le Bénin, l'Afrique subsaharienne et l'Afrique de l'Ouest.

| Nom de la variable | Définition | Unité | Source confirmée | Période | Géographie |
|--------------------|------------|-------|------------------|---------|------------|
| Year | Année de référence pour les données FDI | Aucune| UNCTAD | 1990-2024 | Bénin, Afrique subsaharienne, Afrique de l'Ouest |
| Economy Label | Nom du pays ou de la région (renommé "Country") | Aucune | UNCTAD | 1990-2024 | Bénin, Afrique subsaharienne, Afrique de l'Ouest |
| Flow Label | Type de FDI : Flux annuel) ou Stock (cumulé) | Aucune | UNCTAD | 1990-2024 | Bénin, Afrique subsaharienne, Afrique de l'Ouest |
| Direction Label | Direction du FDI : Inward (entrant) ou Outward (sortant) | Aucune | UNCTAD | 1990-2024 | Bénin, Afrique subsaharienne, Afrique de l'Ouest |
| US$ at current prices in millions | Valeur du FDI en dollars courants | Millions USD | UNCTAD | 1990-2024 | Bénin, Afrique subsaharienne, Afrique de l'Ouest |
| Percentage of total world | Part du FDI dans le total mondial | Pourcentage (%) | UNCTAD | 1990-2024 | Bénin, Afrique subsaharienne, Afrique de l'Ouest |

## 4. Fichier : gdp_data.csv
**Description générale** : Données sur le Produit Intérieur Brut (PIB) réel total et par habitant, incluant taux de croissance annuels, pour le Bénin et l'Afrique subsaharienne.

| Nom de la variable | Définition | Unité | Source confirmée | Période | Géographie |
|--------------------|------------|-------|------------------|---------|------------|
| Year | Année de référence pour les données PIB | Aucune | UNCTAD | 1970-2024 | Bénin, Afrique subsaharienne |
| Country | Nom du pays ou de la région | Aucune | UNCTAD | 1970-2024 | Bénin, Afrique subsaharienne |
| GDP_Current | PIB total en dollars courants (Produit intérieur brut réel : total) | Millions USD | UNCTAD | 1970-2024 | Bénin, Afrique subsaharienne |
| GDP_Current_PerCapita | PIB par habitant en dollars courants (Produit intérieur brut réel : par habitant) | USD par personne | UNCTAD | 1970-2024 | Bénin, Afrique subsaharienne |
| GDP_Constant | PIB total en dollars constants (base 2015, ajusté pour inflation) | Millions USD | UNCTAD | 1970-2024 | Bénin, Afrique subsaharienne |
| GDP_Constant_PerCapita | PIB par habitant en dollars constants | USD par personne | UNCTAD | 1970-2024 | Bénin, Afrique subsaharienne |   

# Légendes des sigles

| Sigle | Définition complète | Description |
|-------|---------------------|-------------|
| UN WPP | United Nations World Population Prospects | Programme des Nations Unies pour les projections démographiques, fournissant des données sur population, natalité, mortalité et migrations. |
| UNDP | United Nations Development Programme | Programme des Nations Unies pour le développement, source des indicateurs sociaux comme HDI, GDI, GII. |
| UNCTAD | United Nations Conference on Trade and Development | Conférence des Nations Unies sur le commerce et le développement, source des données FDI et PIB. |
| HDI | Human Development Index | Indice de développement humain, mesurant santé (espérance de vie), éducation (années de scolarité) et revenu (GNI per capita). |
| GDI | Gender Development Index | Indice de développement par genre, comparant HDI entre hommes et femmes pour évaluer l’égalité de genre. |
| GII | Gender Inequality Index | Indice d’inégalité de genre, mesurant les écarts en santé reproductive, éducation, participation politique et économique entre genres. |
| FDI | Foreign Direct Investment | Investissements directs étrangers, flux (investissements annuels) ou stocks (cumulés), entrants (Inward) ou sortants (Outward). |
| CBR | Crude Birth Rate | Taux brut de natalité, nombre de naissances pour 1000 habitants par an. |
| TFR | Total Fertility Rate | Taux de fécondité total, nombre moyen d’enfants par femme en âge de procréer. |
| LEx | Life Expectancy at Birth | Espérance de vie à la naissance, durée moyenne de vie prévue à la naissance. |
| IMR | Infant Mortality Rate | Taux de mortalité infantile, décès d’enfants de moins d’un an pour 1000 naissances vivantes. |
| Q5 | Under-5 Mortality Rate | Taux de mortalité des moins de 5 ans, décès d’enfants de moins de 5 ans pour 1000 naissances vivantes. |
| GNI | Gross National Income | Revenu national brut, revenu total généré par les résidents d’un pays, ajusté en PPP (parité de pouvoir d'achat). |
| PPP | Purchasing Power Parity | Parité de pouvoir d'achat, ajustement monétaire pour refléter le coût de la vie entre pays. |
